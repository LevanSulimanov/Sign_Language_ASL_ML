```python
sign_language_dataset = []

current_Directory = str(os.getcwd() + '\\Gesture_Image_Data\\')


# just for reference, print # of pixels in image
# borrowed from https://stackoverflow.com/questions/1575625/how-can-i-read-how-many-pixels-an-image-has-in-python
def get_num_pixels(filepath):
    width, height = Image.open(filepath).size
    print(width)
    print(height)
    return width*height



img_labels = []
features = []


# creating column pixel array + label: 50*50 pixels + 1 label
pixel_str = "pixel_"
max_pixels = 2500
dataset_columns = []
#dataset_columns.append("label")
for pixel in range(1, max_pixels+1):
    dataset_columns.append(pixel_str+str(pixel))
    
    
print(dataset_columns)
print(len(dataset_columns))
```

```python
# loop through directories, collect images, convert them into array data, rgb, etc
# looping through directories logic borrowed from https://stackoverflow.com/questions/44112286/python-store-directories-and-filenames-as-dataframe-columns
# image conversion borrowed from https://stackoverflow.com/questions/49649215/pandas-image-to-dataframe
for folder in os.listdir(current_Directory):
    current_folder = os.listdir(current_Directory+folder)
    for image in current_folder:
        # now we go through each image, convert it into rgb 
        image_location = current_Directory+folder + '\\' + image
        colour_img = Image.open(image_location)
        colour_pixels = colour_img.convert("RGB")
        colour_array = np.array(colour_pixels.getdata()).reshape(colour_img.size + (3,))
        indices_array = np.moveaxis(np.indices(colour_img.size), 0, 2)
        #label_array = (str(folder))
        #print(label_array)
        np_image_reshaped = np.dstack((indices_array, colour_array)).reshape((-1, 5))
        #print(pixel_image)
        break
    break
```
